```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('batsman_season_record.csv')
df.head()
```

```python
df.shape
```

```python
plt.style.use('default')
```

```python
df[['batsman','2015']].plot()
```

```python
df = df.set_index('batsman')
```

```python
df.loc['V Kohli'].plot()
```

```python
df.loc['V Kohli'].plot(color='green',linestyle='dashed',marker='o')
df.loc['DA Warner'].plot(color='red',linestyle='dashdot',marker='.')
plt.title('V Kohli V/S DA Warner')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(loc='best')
plt.show()
```

```python
df.loc['V Kohli'].plot(color='green',linestyle='dashed',marker='o')
df.loc['DA Warner'].plot(color='red',linestyle='dashdot',marker='.')
plt.title('V Kohli V/S DA Warner')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(loc='best')
plt.grid()
plt.show()
```
