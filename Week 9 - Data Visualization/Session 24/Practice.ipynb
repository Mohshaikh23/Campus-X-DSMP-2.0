```python
df = pd.read_csv('batter.csv')
df.head()
ipl = df.copy()
```

```python
ipl = ipl.head(25)
plt.figure(figsize=[10,5])
plt.scatter(ipl['avg'],ipl['strike_rate'],s=ipl['runs']/10)
for i in range(ipl.shape[0]):
    plt.text(ipl['avg'].values[i],ipl['strike_rate'].values[i], ipl.batter.values[i])

plt.title('Average v/s Strike rate')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.show()
```

```python
ipl = ipl.head(25)
plt.figure(figsize=[10,5])
plt.scatter(ipl['avg'],ipl['strike_rate'],s=ipl['runs']/10)
for i in range(ipl.shape[0]):
    plt.text(ipl['avg'].values[i],ipl['strike_rate'].values[i], ipl.batter.values[i])

plt.axhline(ipl.strike_rate.mean(),color='red')
plt.axvline(ipl.avg.mean(),color='green')
plt.title('Average v/s Strike rate')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)

z = xx**2 + yy**2

fig = plt.figure(figsize=[10,8])

ax = plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)

z = xx**2 + yy**2

fig = plt.figure(figsize=[10,8])

ax = plt.subplot(2,2,1,projection='3d')
t= ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(t)

new_z = np.sin(xx) + np.sin(yy)
ax = plt.subplot(2,2,2,projection='3d')
t = ax.plot_surface(xx,yy,new_z,cmap='viridis')
fig.colorbar(t)

ax = plt.subplot(2,2,3)
t = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(t)

ax = plt.subplot(2,2,4)
t = ax.contourf(xx,yy,new_z,cmap='viridis')
fig.colorbar(t)

plt.show()
```

```python
df= pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
df.head()
```

```python
temp= df.groupby(['overs','ballnumber'],as_index=False)['total_run'].sum()
temp
```

```python
temp_df = df[df['ballnumber'].isin([1,2,3,4,5,6]) & (df['batsman_run']==6)]
```

```python
grid = temp_df.pivot_table(index='overs',
                   columns='ballnumber',
                   values='batsman_run',
                   aggfunc='count')
```

```python
plt.figure(figsize=[10,5])
plt.imshow(grid.T)
plt.colorbar()
plt.xlabel('Overs')
plt.ylabel('Balls')
plt.xticks(grid.index.unique(),list(range(1,21)))
plt.yticks(np.arange(0,6),list(range(1,7)))
plt.show()
```
