### **Problem-1:** Make a subplots which have 2 plots.

- For the first chart, draw a scatter plot "*Monitored Cap.(MW)*" vs "*Total Cap. Under Maintenance (MW)*" of top 5 most frequent power stations. Then draw the lines which indicate the average values of these two columns. Change the colors according to the names of the Power Stations.
- For the second chart, draw a scatter plot "*Monitored Cap. (MW)*" vs "*Actual(MU)*" of the top 5 most frequent power stations. Also draw the lines which indicates the average values of these two columns. Change the colors according to the names of the Power Stations.


````python
# code here
df = pd.read_csv('https://rb.gy/gsmddu')
df.head()
````

````python
encoder = LabelEncoder()
df['Power_station_num'] = encoder.fit_transform(df['Power Station'])
````

````python
ent = df['Power Station'].value_counts().head().index.tolist()
temp = df[df['Power Station'].isin(ent)]
````

````python
fig,ax = plt.subplots(1,2, figsize=[10,5])

ax[0].scatter(temp['Monitored Cap.(MW)'],
              temp['Total Cap. Under Maintenace (MW)'],
             c=temp['Power_station_num'])
ax[0].axvline(temp['Monitored Cap.(MW)'].mean(),color='red')
ax[0].axhline(temp['Total Cap. Under Maintenace (MW)'].mean(),color='blue')
ax[0].set_xlabel('Monitored Cap.(MW)')
ax[0].set_ylabel('Total Cap. Under Maintenace (MW)')
ax[0].set_title('Monitored Cap V/s Total Cap. Under Maintenace')


ax[1].scatter(temp['Monitored Cap.(MW)'],
              temp['Actual(MU)'],
             c=temp['Power_station_num'])
ax[1].axvline(temp['Monitored Cap.(MW)'].mean(),color='red')
ax[1].axhline(temp['Actual(MU)'].mean(),color='blue')
ax[1].set_xlabel('Monitored Cap.(MW)')
ax[1].set_ylabel('Actual(MU)')
ax[1].set_title('Monitored Cap V/s Actual(MU)')


plt.show()
````

### **Problem-2:** Draw a 3D Scatter plot between "*Monitored Cap.(MW)*", "*Total Cap. Under Maintenace (MW)*" and "*Forced Maintanence(MW)*"

````python
# code here
fig = plt.figure(figsize=[10,8])

ax = plt.subplot(projection='3d')

ax.scatter3D(df['Monitored Cap.(MW)'],
             df['Total Cap. Under Maintenace (MW)'],
             df['Forced Maintanence(MW)'],
             c = df['Power_station_num'])

ax.set_title("Monitored Cap.(MW) v/s Total Cap. Under Maintenace (MW) v/s Forced Maintanence(MW)")
ax.set_xlabel('Monitored Cap.(MW)')
ax.set_ylabel('Total Cap. Under Maintenace (MW)')
ax.set_zlabel('Forced Maintanence(MW)')

plt.show()
````

### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

````python
# code here

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)

z = np.abs(xx) - np.abs(yy)

fig = plt.figure(figsize=[10,8])

ax = plt.subplot(projection='3d')
t = ax.plot_surface(xx,yy,z,cmap='viridis')
plt.colorbar(t)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
````

### `Problem-6` Use Pandas plot functions

* Line plot of closing value of top 5 Stocks in Year 2020.
* Take top 5 stocks based on total turnover in Year 2020

````python
# code here
df= pd.read_csv("nifty-50 - nifty-50.csv")
df.Date = pd.to_datetime(df.Date)
df = df.set_index('Date')
````

````python
top5 = df.groupby('Symbol')['Turnover'].sum().sort_values(ascending=False).head().index.tolist()
top5_data = df[df.Symbol.isin(top5)].loc['2020']
````

````python
for name,group in top5_data.groupby('Symbol'):
    plt.plot(group.index, group['Close'],label=name)

plt.legend()
plt.show()
````

### `Problem 9:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = sin(\sqrt{x^2 + y^2})$$


````python
# code here
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)

z = np.sin(np.sqrt(xx**2 + yy**2))

fig = plt.figure(figsize=[10,5])
ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
````

````python
fig = plt.figure()
ax = plt.subplot()

ax.contour(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
````

````python
fig = plt.figure()
ax = plt.subplot()

ax.contourf(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
````

### `Problem 10:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = tan(\log_2({x^2 + y^2})$$

````python
# code here
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)

z = np.tan(np.log2(xx**2 + yy**2))

fig = plt.figure()
ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
````
