```python
df3 = pd.DataFrame({'Age':[25,30,35,40,45,50,22,38,35,42,43,51],
                   'Sex':['F','M','F','M','F','M','M','M','M','F','M','M'],
                   'Country':['US','UK','CAN','AUS','NZ','US','US','UK','CAN','AUS','NZ','UK']})
df4 = df3.pivot_table(index=['Country','Sex'],values='Age',aggfunc='mean',margins=True)
print(df4.unstack('Country'))
print(df4.shape)
```

```python
arrays = [np.array(['bar','bar','baz','baz','foo','foo','qux','qux']),
         np.array(['one','two','one','two','one','two','one','two'])]
df= pd.DataFrame(np.random.randn(8,2), index=arrays)
df.index.set_names(['first','second'],inplace=True)
df = df.rename_axis(['A','B'],axis=0)
df.groupby(level='A').agg({0:['mean','min'],1:['max','sum']})
```
